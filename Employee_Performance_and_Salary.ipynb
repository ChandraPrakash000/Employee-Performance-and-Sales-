## Q21.Add a new column Age Group with values "Young", "Middle-aged", or "Senior" based on Age

```python
import numpy as np

conditions=[
    (df['Age']<30),
    (df['Age']>30) & (df['Age']<60),
    (df['Age']>=60)
]
choices=['Young','Middle-aged','Senior']

df['Age Group']=np.select(conditions,choices,default='unknown')
df.head()
```

## 46. Create a second dataset with Department and Budget. Perform an inner join with the original dataset

```python
dept_budget = pd.DataFrame({
    'Department': ['HR', 'Engineering', 'Marketing'],
    'Budget': [200000, 500000, 300000]
})

df = pd.merge(df, dept_budget, on='Department', how='inner')
```

## 47. Merge the dataset with another dataset that includes employee ID and their recent Projects

```python
recent_projects = pd.DataFrame({
    'ID': [101, 102, 103, 104],
    'Recent Project': ['Recruitment Drive', 'AI Development', 'Ad Campaign', 'System Upgrade']
})

df = pd.merge(df, recent_projects, on='ID', how='left')
```

## 48. Perform a left join with a dataset containing Department and Team Lead details

```python
team_leads = pd.DataFrame({
    'Department': ['HR', 'IT', 'Sales'],
    'Team Lead': ['Eve', 'Frank', 'Grace']
})

df = pd.merge(df, team_leads, on='Department', how='left')
```

## 49. Create a DataFrame with employee ID and Project Count, then merge it and calculate average Salary per project

```python
project_count = pd.DataFrame({
    'ID': [101, 102, 103, 104],
    'Project Count': [2, 4, 3, 2]
})

df = pd.merge(df, project_count, on='ID', how='left')
df['Avg Salary per Project'] = df['Salary'] / df['Project Count']
```

## 50. Calculate the percentage increase in Salary compared to the average Salary in their Department

```python
avg_salary_by_dept = df.groupby('Department')['Salary'].transform('mean')
df['% Increase vs Dept Avg'] = ((df['Salary'] - avg_salary_by_dept) / avg_salary_by_dept) * 100
```
